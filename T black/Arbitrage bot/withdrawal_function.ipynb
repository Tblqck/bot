```python
import requests

def fetch_withdrawal_fees_binance():
    url = 'https://api.binance.com/api/v3/assetDetail'
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        for asset, asset_info in data.items():
            if 'withdrawFee' in asset_info and 'withdrawStatus' in asset_info:
                withdrawal_fee = asset_info['withdrawFee']
                withdraw_status = asset_info['withdrawStatus']
                print(f"{asset} Withdrawal Fee: {withdrawal_fee}, Withdraw Status: {withdraw_status}")
        print("------------------------")
    else:
        print(f"Error occurred while fetching data from Binance. Status Code:", response.status_code)

if __name__ == '__main__':
    fetch_withdrawal_fees_binance()
```

```python
import requests

def fetch_withdrawal_fees_kraken():
    url = 'https://api.kraken.com/0/public/AssetPairs'
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()['result']
        for trading_pair, pair_info in data.items():
            withdrawal_fee = pair_info['fees'][0][1]
            print(f"{trading_pair} Withdrawal Fee: {withdrawal_fee}")
        print("------------------------")
    else:
        print(f"Error occurred while fetching data from Kraken. Status Code:", response.status_code)

if __name__ == '__main__':
    fetch_withdrawal_fees_kraken()
```
