```python
import requests

def fetch_matic_historical_data():
    base_url = "https://api.mexc.com"
    endpoint = "/api/v3/klines"
    symbol = "MATICUSDT"  # Trading pair symbol
    interval = "1m"       # Kline interval (1 day)
    limit = 100           # Number of klines to fetch

    params = {
        "symbol": symbol,
        "interval": interval,
        "limit": limit
    }

    response = requests.get(base_url + endpoint, params=params)

    if response.status_code == 200:
        data = response.json()
        print("Timestamp, Open, High, Low, Close, Volume")
        for kline in data:
            timestamp = kline[0]
            open_price = kline[1]
            high_price = kline[2]
            low_price = kline[3]
            close_price = kline[4]
            volume = kline[5]
            print(f"{timestamp}, {open_price}, {high_price}, {low_price}, {close_price}, {volume}")
    else:
        print(f"Error occurred. Status Code: {response.status_code}")

if __name__ == "__main__":
    fetch_matic_historical_data()
```
