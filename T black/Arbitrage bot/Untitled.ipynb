```python
import requests

def fetch_matic_usdt_ticker():
    base_url = "https://api-cloud.bitmart.com"
    endpoint = "/spot/quotation/v3/tickers"
    url = base_url + endpoint

    response = requests.get(url)
    data = response.json()

    matic_usdt_ticker = None

    for ticker in data['data']:
        if ticker[0] == "MATIC_USDT":
            matic_usdt_ticker = {
                "symbol": ticker[0],
                "last": ticker[1],
                "v_24h": ticker[2],
                "qv_24h": ticker[3],
                "open_24h": ticker[4],
                "high_24h": ticker[5],
                "low_24h": ticker[6],
                "fluctuation": ticker[7],
                "bid_px": ticker[8],
                "bid_sz": ticker[9],
                "ask_px": ticker[10],
                "ask_sz": ticker[11],
                "ts": ticker[12]
            }
            break

    return matic_usdt_ticker

matic_usdt_ticker = fetch_matic_usdt_ticker()
print(matic_usdt_ticker)
```

```python
import requests

def fetch_historical_klines(symbol, step, before, after, limit):
    base_url = "https://api-cloud.bitmart.com"
    endpoint = "/spot/quotation/v3/lite-klines"
    url = base_url + endpoint

    params = {
        "symbol": symbol,
        "step": step,
        "before": before,
        "after": after,
        "limit": limit
    }

    response = requests.get(url, params=params)
    data = response.json()

    return data['data']

symbol = "MATIC_USDT"
step = 15  # K-Line step (in minutes)
before = None  # Specify the timestamp before which data is needed
after = None   # Specify the timestamp after which data is needed
limit = 100    # Number of data points to retrieve

historical_data = fetch_historical_klines(symbol, step, before, after, limit)
for data_point in historical_data:
    print(data_point)
```

```python
import requests

def fetch_ticker(symbol):
    base_url = "https://openapi.digifinex.com"
    endpoint = "/v3/ticker"
    url = f"{base_url}{endpoint}?symbol={symbol}"

    response = requests.get(url)
    data = response.json()

    if "ticker" in data:
        ticker_data = data["ticker"][0]
        return ticker_data
    else:
        return None

symbol = "matic_usdt"
ticker_data = fetch_ticker(symbol)

if ticker_data:
    print(ticker_data)
else:
    print("Ticker data not found.")
```

```python
import requests

def fetch_candlestick_data(symbol, period, start_time=None, end_time=None):
    base_url = "https://openapi.digifinex.com"
    endpoint = "/v3/kline"
    url = f"{base_url}{endpoint}?symbol={symbol}&period={period}"

    if start_time:
        url += f"&start_time={start_time}"
    if end_time:
        url += f"&end_time={end_time}"

    response = requests.get(url)
    data = response.json()

    if "data" in data:
        candlestick_data = data["data"]
        return candlestick_data
    else:
        return None

symbol = "matic_usdt"
period = "1"  # 1-minute candles
candlestick_data = fetch_candlestick_data(symbol, period)

if candlestick_data:
    print(candlestick_data)
else:
    print("Candlestick data not found.")
```
